## Step 2: Review the BAD Charts

Below are two deliberately poor visualizations.  
Your task: Carefully review each and take notes on what you think is wrong, misleading, or just poorly designed.  
Think about chart type, colors, scales, labeling, and whether the chart tells the truth.  


```python
plt.figure(figsize=(8,5))
subset = coffee_df.groupby("Year")["Sales"].sum().reset_index()

# Wrong variable for title, truncated y-axis, misleading color & framing
plt.plot(subset["Year"], subset["Sales"], color="red", linewidth=4, marker="o")
plt.ylim(subset["Sales"].min() * 0.98, subset["Sales"].min() * 1.02)  # extreme truncation
plt.title("Shocking Collapse in Profits!", fontsize=16, fontweight="bold", color="darkred")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

### Write your notes about bad chart #1 here:
- Emotionally charged chart title
- Y axis starts as 2.90 instead of 0
- Axes are not labelled 
- X axis uses non-standard time format (0.5 of a year? )
- Y axis appears truncated: graph appears to continue beyond cutoff but is not visible 
- '1e6' label on y-axis indicates possible exponent used? Could be sheisty math to manipulate data. 
- Chart label says 'profits' but y-axis data is being pulled from sales column 

```python
plt.figure(figsize=(10,6))
avg_sales = coffee_df.groupby("Region")["Sales"].mean().reset_index()

colors = ["red","orange","yellow","green","blue","purple"]

plt.bar(avg_sales["Region"], avg_sales["Sales"], 
        color=colors, edgecolor="black", linewidth=3, hatch="/", alpha=0.8)
plt.title("!!! INCREDIBLE SALES DATA !!!", fontsize=20, fontweight="bold", color="purple", backgroundcolor="yellow")
plt.grid(True, which='both', color='black', linestyle='-', linewidth=1.5)
plt.xlabel("Region of the WORLD!!!!", fontsize=16, fontweight="bold", color="darkgreen")
plt.ylabel("MONEY $$$$$", fontsize=16, fontweight="bold", color="darkblue")
plt.xticks(rotation=45, fontsize=14, fontweight="bold", color="orange")
plt.yticks(fontsize=14, fontweight="bold", color="orange")
plt.show()
```

### Write your notes about bad chart #2 here:
- Emotionally charged chart label
- Vague and unprofessional labels on axes 
- Busy chart design and color scheme decreases chart legibility 
- x-axis is unclear (what does 'region of the world' mean? what do these cardinal directions apply to?)
- y-axis is not clear (money for what? in USD? are these sales numbers or something else?)

```python
subset = coffee_df.groupby("Year")["Profit"].sum().reset_index()
subset

# cleaner look at what's going on - 5 rows with annual profit by year 
```

```python
# want to get an ugly first pass of a chart with no funny business 

plt.figure(figsize=(10,6))
plt.ylim(0, subset["Profit"].max() * 1.1)  # set the y axes to start at zero and end 10% above max y value 

plt.xlabel("Year")
plt.ylabel("Profit ($)") # The README states values are in millions, but the magnitudes suggest raw dollars. There's no way they sell that much.
plt.title("Annual Profit (2020-2024)")

plt.xticks(subset["Year"]) # want to get rid of half-years
 
plt.plot(subset["Year"], subset["Profit"], marker="o")
```

## Chart #1 Reflection 

The original chart utilized alarmist and emotionally manipulative titling, plotted sales data but labeled it as profits, truncated the y-axis leading to exaggeration of the profit dip, used emotionally evocative colors (red), and displayed half-years on the x-axis. These issues have been corrected in the updated chart: the title is neutral and accurate, the correct variable (Profit) is plotted and labeled, the y-axis starts at zero, a neutral color is used, and the x-axis shows only actual years.

```python
# want to see the average sales for each region for chart #2

region_avg_sales = coffee_df.groupby("Region")["Sales"].mean().reset_index()
region_avg_sales
```

```python
# sales numbers are all very close to each other numerically but this is obfuscated by chart junk in original chart 
# making a clean bar graph to see how the sales stack up visually without the chartjunk 

plt.Figure(figsize=(8,5))
plt.bar(region_avg_sales["Region"], region_avg_sales["Sales"])

plt.xlabel("Region")
plt.ylabel("Average Sales ($)")
plt.title("Average Sales by Region")

plt.show()
```

## Chart #2 Reflection

The original chart used an emotionally charged title, vague and unprofessional axis labels ("MONEY $$$$$", "Region of the WORLD!!!!"), distracting colors that implied meaningful differences between regions, and unnecessary chartjunk (hatching, heavy gridlines, bold formatting throughout). The corrected chart uses a neutral descriptive title, clear axis labels, and a single color for all bars. This reveals the actual story: all four regions perform nearly identically, with average sales ranging from approximately \$12,450 to \$12,980 (a difference of only about 4%).##

**Note on units:** The README states that sales values are in millions, but the data magnitudes suggest otherwise. A single product category in one region for one month shows values around \$14,000–\$19,000. If these were millions, that would represent billions in monthly revenue for a single product line, which is implausible for a coffee chain. I've labeled the axes as "Profit (\$)" and "Average Sales (\$)" rather than "Profit (in millions)" and "Sales (in millions)" based on this observation.
